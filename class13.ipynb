```python
# importing matplotlib module and numpy
from matplotlib import pyplot as plt 
import numpy as np

# x-axis values 
x = [1,2,3,4,5,6] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2,2] 
  
# Function to plot 
plt.plot(x, y) 
  
# function to show the plot 
plt.show()
```

Strange plot, but not bad for a start...  Maybe a bar plot makes more sense for this data?

```python
# Function to plot 
plt.bar(x, y) 
  
# function to show the plot 
plt.show()
```

Or, we could create a histogram (binned data):

```python
# Function to plot histogram 
plt.hist(y) 
  
# Function to show the plot 
plt.show() 
```

Pretty boring histogram, but easy!  We can also make a 2-D scatter plot of this data:

```python
# Function to plot scatter 
plt.scatter(x, y) 
  
# function to show the plot 
plt.show() 
```

There are many options that we can use for point style and color:
![image](img/options1.png)
![image](img/options2.png)


Let's try one...

```python
# Function to plot scatter 
plt.scatter(x, y,c="g", alpha=0.5, marker='s') 

# function to show the plot 
plt.show() 
```

Honestly, all of those plots are terrible!  At the very least they need axis lables!

```python
# Function to plot 
plt.scatter(x, y) 
  
# Adding Title
plt.title("Our First Plot Title")
  
# Labeling the axes
plt.xlabel("X-axis [units]")
plt.ylabel("Y-axis [units]")
  
# function to show the plot 
plt.show()
```

We could draw multiple plots on the same "canvas"

```python
plt.scatter(x, y) 
  
# function to show the plot 
plt.show()
  
plt.plot(x, y)
  
# function to show the plot 
plt.show()
```

We can also use subplot- it is fine for one plot, but also gives flexibility for displaying more plots: for example to draw 2 plots next to each other...

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('A single plot')
```

Let's see how to draw two plots next to each other - to obtain side-by-side subplots, pass parameters 1, 2 for one row and two columns.

```python
# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x)
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
```

## ** Plotting the Iris Dataset **: 

Let's now plot some data!  In a past HW, we split the Iris data into three class files.  First we need to load our datasets.  It is easy with loadtxt:

```python
np_list_virginica= np.loadtxt("Iris-virginica.dat")
np_list_versicolor= np.loadtxt("Iris-versicolor.dat")
print(np_list_virginica)
```


That worked!  

Remember,the Iris is a flower and our datasets has 4 collums of data:  
1. sepal length in cm
2. sepal width in cm
3. petal length in cm
4. petal width in cm

![Image](img/index.jpg)

Lets make a histogram of sepal length for Iris-virginica:


```python
histo_data=np_list_virginica[:,0]
print(histo_data)
```

That worked, let's plot it:

```python
plt.hist(histo_data) 
```

Of course, we don't need the independent step of making the 1D-array:

```python
plt.hist(np_list_virginica[:,0],label="Virginica")
plt.hist(np_list_versicolor[:,0],label="Versicolor")
# Adding Title
plt.title("Iris Class Characteristics")
  
# Labeling the axes
plt.xlabel("Septal length (cm)")
plt.ylabel("Count/bin")
plt.legend()  
```

I like the UVA color theme!

Also, this is very interesting - the distribution of septal length, looks different for these two classes of Iris!  So, this variable provides some discriminating power between the Virginica and Versicolor classes of Iris.   

It is trivial to save a figure to a file, but we need to make our plot a figure first.

```python
plt.figure()  #Need this before we are able to save a plot to file!!!

plt.hist(np_list_virginica[:,0],label="Virginica")
plt.hist(np_list_versicolor[:,0],label="Versicolor")
# Adding Title
plt.legend()
plt.title("Iris Class Characteristics")
  
# Labeling the axes
plt.xlabel("Septal length (cm)")
plt.ylabel("Count/bin")
 
plt.savefig('septal_length.png')
```
